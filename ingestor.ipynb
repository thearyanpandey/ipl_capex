```python
import pandas as pd
import sqlite3
```

```python
matches_df = pd.read_csv('matches.csv')
deliveries_df = pd.read_csv('deliveries.csv')
# matches_df[matches_df['method'].notna()]
deliveries_df.head(5)
# print("Data loaded successfully")
```

```python
print("---Missing data in Matches---")
print(matches_df.isnull().sum())
# matches_df[
#     matches_df["city"].isna()
# ]
# print("\n")
```

```python
print("----Missing data in Deliveries---")
print(deliveries_df.isnull().sum())
```

```python
#-----Cleaning the matches dataset-----
#1. Filling missing cities with Venue name
matches_df['city'] = matches_df['city'].fillna(matches_df['venue'])

#2. Drop the 5 matches that had no result/winner 
matches_df = matches_df.dropna(subset=['winner', 'player_of_match'])

#3. Drop columns which we don't need for Financial/ROI model
matches_df = matches_df.drop(['method', 'umpire1', 'umpire2'], axis=1)
```

```python
#----Cleaning the deliveries dataset----
#1. Since SQL is not good with NaN we'll replace those with NONE
deliveries_df = deliveries_df.fillna('None')
```

```python
print("Matches data cleaned")
print(matches_df.isnull().sum())
print("\n")
print(deliveries_df.isnull().sum())
```

```python
conn = sqlite3.connect('ipl_master_database.db')

matches_df.to_sql('matches_table', conn, index=False, if_exists='replace')
deliveries_df.to_sql('deliveries_table', conn, index=False, if_exists='replace')

conn.close()
```
